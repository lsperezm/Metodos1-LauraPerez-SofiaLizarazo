```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v_0=1
theta=1
R=8.314
V=1
N=1
```

```python
U_=np.arange(1,100,0.5)
```

```python
def EC1(U):
    ln=(U*V)/((N**2)*R*theta*v_0)
    S=N*R*np.log(ln)
    return S
```

```python
S_1=EC1(U_)
```

```python
plt.plot(U_,S_1,label="Ec1")
plt.xlabel("Energía Interna")
plt.ylabel("Entropía")
plt.legend()
```

```python
def EC4(U):
    r=(R/theta)**(1/2)
    exp=(-(U*V)/(N*R*theta*v_0))
    n=(N*U)**(1/2)
    
    S=r*n*np.exp(exp)
    return S
```

```python
S_4=EC4(U_)
```
