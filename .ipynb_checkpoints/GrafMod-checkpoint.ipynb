```python
%matplotlib inline

from __future__ import division

import quantities as pq
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
```

```python
K_b=1.3806488e-23
T_1=90
T_2=300
T_3=800
T_4=5000
h=6.62607015e-34
c=2.99e8
```

$\frac{(2*\pi*(c^2)*h)}{((e^(\frac{h*c}{(k_b*T*lam)}) - 1)}$

```python
x_4=np.arange(0.01,4.01,0.01)
lam_4=x_4*1e-6
x_3=np.arange(0.01,20.01,0.01)
lam_3=x_3*1e-6
x_2=np.arange(0.01,50.01,0.01)
lam_2=x_2*1e-6
x_1=np.arange(0.01,180.01,0.01)
lam_1=x_1*1e-6
```

```python
def Rad(lam,T):
    energia_luz=h*c/lam
    energia_termica1=K_b*T
    prob1=1/(np.exp(energia_luz/energia_termica1)-1)

    planck1=(2*np.pi*h*c**2/lam**5)*prob1
    
    return planck1
```

```python
plt.plot(lam_1,Rad(lam_1,T_1), label="T=90K")

plt.legend()
plt.xlabel("Longitud de Onda (m)")
plt.ylabel("Intensidad")

```

```python
plt.plot(lam_2,Rad(lam_2,T_2), label="T=300K",color="green")
plt.legend()
plt.xlabel("Longitud de Onda (m)")
plt.ylabel("Intensidad")
```

```python
plt.plot(lam_3,Rad(lam_3,T_3), label="T=800K",color="yellow")

plt.legend()
plt.xlabel("Longitud de Onda (m)")
plt.ylabel("Intensidad")
```

```python
plt.plot(lam_4,Rad(lam_4,T_4), label="T=5000K",color="orange")

plt.legend()
plt.xlabel("Longitud de Onda (m)")
plt.ylabel("Intensidad")
```
